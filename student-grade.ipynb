```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import  mean_absolute_error , mean_squared_error , r2_score
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

data=pd.read_csv("student_performance.csv").head(500)
data
```

```python
x=data[["weekly_self_study_hours"]]
y=data["total_score"]
model=LinearRegression()
model.fit(x,y)
pred=model.predict(x)
mae=mean_absolute_error(y,pred)
mse=mean_squared_error(y,pred)
rmse=np.sqrt(mse)
r2=r2_score(y,pred)

print(f"MAE : {mae:.4f}")
print(f"MSE : {mse:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R^2 SCORE : {r2:.4f}")
```

```python
plt.bar(x=data["grade"] ,height=data["total_score"] ,color="green" ,edgecolor="black" )
plt.plot(data["grade"] , pred ,color="blue" )
plt.title("student grade vs marks....")
plt.xlabel("Grade")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
